```python
#since age is numerical value we will use median here
median = df['Age'].median()
df['Age']=df['Age'].fillna(median)

df['Age'].isna().sum() #tells null values after cleaning
```

```python
#embarked is categorical data we will use mode
mode = df['Embarked'].mode()[0]
df['Embarked']=df['Embarked'].fillna(mode)

df['Embarked'].isna().sum() #null values after cleaning
```

```python
df[['Age','Cabin','Embarked']].isna().sum()
```

```python
#cabin is categorical but has over 70% null values , we can't use mode. 
df['Has_cabin']=df['Cabin'].notna().astype(int)
df.drop('Cabin', axis=1, inplace=True)
```

```python
#univariate analysis
# for catgeorical : bar charts , pie charts
#for numericals: histograms , kdeplot
```

```python
print("Analysing categorical features: ")
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Univariate Analysis of Categorical Features', fontsize=16)

sns.countplot(ax=axes[0,0],x='Survived',data=df).set_title("Survival Dist.")
sns.countplot(ax=axes[0,1],x='Pclass',data=df).set_title("Passenger Class Dist.")
sns.countplot(ax=axes[0,2],x='Sex',data=df).set_title("Gender Dist.")
sns.countplot(ax=axes[1,0],x='Embarked',data=df).set_title("Port of Embarrked")
sns.countplot(ax=axes[1,1], x='SibSp',data=df).set_title("Spouses")
sns.countplot(ax=axes[1,2],x='Parch',data=df).set_title("Parents and children")

plt.tight_layout(rect=[0,0,1,0.96])
plt.show()
```

```python
print("Analysing numerical features: ")
fig, axes = plt.subplots(1, 2, figsize=(18, 6))
fig.suptitle('Univariate Analysis of Numerical Features', fontsize=16)

sns.histplot(ax=axes[0],x='Age',kde=True,data=df,bins=30).set_title("Age Dist.")
sns.histplot(ax=axes[1],x='Fare',kde=True,data=df,bins=30).set_title("Fare Dist.")

plt.show()
```

```python
'''
Bivarate Analysis:
Cat v/s Num : bar plot 
Cat v/s Cat: Stacked bar
Num v/s Num: Scatter plot
'''
```

```python
print("Bivariate Analysis: Feature vs. Survival")

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Bivariate Analysis with Survival', fontsize=16)

sns.barplot(ax=axes[0, 0], x='Pclass', y='Survived', data=df).set_title('Survival Rate by Pclass')
sns.barplot(ax=axes[0, 1], x='Sex', y='Survived', data=df).set_title('Survival Rate by Sex')
sns.barplot(ax=axes[1, 0], x='Embarked', y='Survived', data=df).set_title('Survival Rate by Port')
sns.barplot(ax=axes[1, 1], x='Has_cabin', y='Survived', data=df).set_title('Survival Rate by Cabin Availability')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
#facetgrid for age v/s survived
g = sns.FacetGrid(df, col='Survived', height=6)
g.map(sns.histplot, 'Age', bins=20, kde=True)
plt.suptitle('Age Distribution by Survival Status', y=1)
plt.show()
```

```python
#outlier analysis for fare
plt.figure(figsize=(8, 8))
sns.boxplot(y='Fare', data=df)
plt.title('Box Plot of Ticket Fare')
plt.ylabel('Fare')
plt.show()
```

```python
'''
Feature Engineering: 
1. Combining Features: a new feature by combining others
2. Binning : Numerical to categorical or vice versa
3. Extracting from text : Takinf some part out of whole cat values
'''
```

```python
#combining
df['Family Size']= df['Parch']+df['SibSp']+1
df['IsAlone']=0 #has a family
df.loc[df['Family Size']==1,'IsAlone']=1 

df.head()
```

```python
#extracting from text
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)

print("Extracted Titles:")
df['Title'].value_counts()
```

```python
df['Title'] = df['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')

df['Title'] = df['Title'].replace('Mlle', 'Miss')
df['Title'] = df['Title'].replace('Ms', 'Miss')
df['Title'] = df['Title'].replace('Mme', 'Mrs')

df.head(10)
df['Title'].value_counts()
```

```python
#Multivariate Analysis
sns.catplot(x='Pclass', y='Survived', hue='Sex', data=df, kind='bar', height=6, aspect=1.5)
plt.title('Survival Rate by Pclass and Sex')
plt.ylabel('Survival Probability')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.violinplot(x='Sex', y='Age', hue='Survived', data=df, split=True, palette={0: 'blue', 1: 'orange'})
plt.title('Age Distribution by Sex and Survival')
plt.show()
```

```python
# Correlation Heatmap for numerical features
plt.figure(figsize=(14, 10))
numeric_cols = df.select_dtypes(include=np.number)
matrix = numeric_cols.corr()
sns.heatmap(matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
